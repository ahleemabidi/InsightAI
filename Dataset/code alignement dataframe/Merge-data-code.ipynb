```python

import pandas as pd
file_path = r'C:\Users\user\Desktop\stageeee\dataset\Commits_functions.csv'
Commits_functions = pd.read_csv(file_path) 
Commits_functions
```

```python

# Define the columns that represent functions
function_columns = Commits_functions.columns[2:-1]  # Exclude 'commit', 'message', and 'ID'

```

```python

# Create the 'functions' column
Commits_functions['functions'] = Commits_functions.apply(lambda row: [col for col in function_columns if row[col] == 1], axis=1)
```

```python


# Drop the original function columns if desired
Commits_functions = Commits_functions.drop(columns=function_columns)
```

```python
import pandas as pd

# Define the function
def delete_column(df, column_name):
    """
    Delete a column from a DataFrame.

    Parameters:
    df (pd.DataFrame): The DataFrame from which to delete the column.
    column_name (str): The name of the column to delete.

    Returns:
    pd.DataFrame: The DataFrame with the column removed.
    """
    return df.drop(columns=[column_name])
merged = delete_column(merged, 'Message')
```

```python
merged = delete_column(merged, 'Title')
```

```python
merged
```

```python
# Define the desired column order
desired_order = ['ID', 'Date', 'commit', 'message', 'functions', 'User', 'Author', 'Created At', 'Updated At', 'Labels', 'State','Duration']


# Reorder the columns
merged = merged.reindex(columns=desired_order)

merged
```
